```python
import numpy as np
import math
import qiskit
from math import sqrt

from qiskit import QuantumCircuit
from qiskit import execute, Aer, BasicAer
from qiskit.tools.visualization import circuit_drawer
from qiskit.tools.visualization import plot_histogram
from qiskit.tools.visualization import plot_bloch_multivector
from qiskit.quantum_info import Statevector
from qiskit import transpile
```

Soal matematis nomer 5

```python
qc = QuantumCircuit(3)        # Buat sirkuit kuantum dengan 4 qubit: q2, q1, q0
qc.h(0)
qc.h(2)
qc.barrier()
qc.y(0)
qc.z(1)
qc.x(2)
qc.barrier()
qc.cx(0,1)
qc.y(2)
qc.barrier()
qc.z(1)
qc.swap(0,2)
qc.barrier()

qc.ccx(2,1,0)


qc.draw('mpl')
```

```python
# Keadaan Akhir dari qc
psi_1=Statevector(qc)
psi_1.draw('latex')
```

Soal Program 1.a

```python
qc1 = QuantumCircuit(3)
qc1.h([0,1,2])
qc1.barrier()
qc1.x(0)
qc1.cx(1,2)
qc1.y(2)
qc1.barrier()
qc1.ccx(0,1,2)
qc1.x(0)
qc1.y(1)
qc1.barrier()
qc1.z(1)
qc1.swap(0,2)
qc1.barrier()

qc1.draw('mpl')
```

Soal program 1.b


```python
qc2 =QuantumCircuit(3)
qc2.h([0,2])
qc2.x(1)
qc2.barrier()
qc2.cx(0,1)
qc2.z([0,2])
qc2.barrier()
qc2.z([0,2])
qc2.cx(0,1)
qc2.barrier()
qc2.h([0,2])
qc2.x(1)
qc2.draw('mpl')
```

Soal program 1.c


```python
qc3 = QuantumCircuit(3)
qc3.h([0,1,2])
qc3.barrier()
qc3.x(1)
qc3.h(2)
qc3.ccx(0,1,2)
qc3.x(1)
qc3.h(2)
qc3.barrier()
qc3.draw('mpl')
```
